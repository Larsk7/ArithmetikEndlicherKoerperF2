```python
import time as t
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Bestimme die Länge der Tupel, die auf Irreduzibilität geprüft werden sollen
length = 9

# Erstelle eine Liste aller Tupel der Länge 'length'
list_of_all_tupels = list(product([0, 1], repeat=length))[4:]

# Variablen zum Zählen, wie viele Ergebnisse (nicht) übereinstimmen
correct_counter, mistake_counter = 0, 0

# Schleife über alle Tupel der Liste
for F in list_of_all_tupels:

    # Wenn Ergebnisse übereinstimmen, 'correct_counter' eins hochzählen
    if is_minpol(F) == cantor_zassenhaus(F):
        correct_counter = correct_counter + 1

    # Wenn Ergebnisse nicht übereinstimmen, 'mistake_counter' eins hochzählen
    else:
        mistake_counter = mistake_counter + 1

# Ausgabe, wie viele Ergebnisse (nicht) übereinstimmen
print(f"Falsch: {mistake_counter}")
print(f"Korrekt: {correct_counter}")
```

```python
# Alle Tupel der Länge 'length' mittels "MinimalPolynoms" überprüfen

# Startzeitpunkt
start1 = t.time()

# Iteration über alle Tupel-Multiplikationen
for F in list_of_all_tupels: 
        A = is_minpol(F)

# Endzeitpunkt
end1 = t.time()
```

```python
# Alle Tupel der Länge 'length' mittels "CantorZassenhaus" überprüfen

# Startzeitpunkt
start2 = t.time()

# Iteration über alle Tupel-Multiplikationen
for F in list_of_all_tupels: 
        B = cantor_zassenhaus(F)

# Endzeitpunkt
end2 = t.time()
```

```python
# Differenz der Zeitpunkte bilden und ausgeben (auf 5 Nachkommastellen)
print(f"MinimalPolynoms:  {(end1 - start1):.5f} Sekunden")
print(f"CantorZassenhaus: {(end2 - start2):.5f} Sekunden")
```
